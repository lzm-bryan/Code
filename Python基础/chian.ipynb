```python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None    
```

```python
node = ListNode(0)
node.val = 5
node.next = None
print(node.val)  # Output: 5
#注释：定义了一个链表节点类ListNode，并创建了一个节点实例，赋值并打印节点的值
```

```python
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node1.next = node2
node2.next = node3
print(node1.next.val)  # Output: 2
while node1 is not None:
    print(node1.val)
    node1 = node1.next  
#注释：遍历链表并打印每个节点的值
```

```python
def example_ternary(x):
    x = 10 if x > 5 else 0
    return x    
#判断x是否大于5，大于则赋值10，否则赋值0
```

```python
example_ternary(7)  # Returns 10
```

```python
def function(item):
    x = [i for i in item if i != 0]
    print(x)
function([1,2,5,0,1])
```
